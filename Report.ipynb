## Project Submission: Collaboration and Competition

### Learning algorithm

This submission solves the Tennis environment. While there are two agents playing against each other, it is possible to simultaneously train both of them. This is equivalent to training both the Actor and the Critic networks from the same replay buffer. The agent used is a Deep Deterministic Policy Gradient agent. 

The Actor consists of two identical networks corresponding to a target copy to which updates are made and a local copy from which actions are selected. Weights are copied from the target to the local using a soft update. The actor networks consist of three fully connected layers where the first and second layers have 128 nodes each. A ReLU function is used between layers. Batch normalization is applied to the output of the first layer. The final output is obtained by applying a Tanh function to the output of the third layer.

The Critic similarly consists of two identical networks (target and local). Each network has three layers where the first layer has 128 nodes while the second layer has 130 nodes (128 + 2 corresponding to size of the action space). ReLU functions are applied out the output of the fully connected layers. However, unlike the Actor networks, there is no Tanh applied on the final output. Weights are transferred from the target to the local networks using the mechanism described for the Actor. 

Both networks are tuned using an Adam optimizer and use an Ornstein-Uhlenbeck process for regularization. 

### Parameters

Buffer size: 100000  
Batch size: 128  
Discount factor: 0.99  
Soft update rate: 0.001  
Actor learning rate: 0.001  
Critic learning rate: 0.001  
OU noise mean: 0  
OU noise theta: 0.15  
OU noise sigma: 0.1
